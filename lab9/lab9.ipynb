```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def system(z, y):
    return z, -np.sin(y)

def InitialСonditions(alpha):
    return 0, alpha
```

```python
def  solve(N, rhs, method, alpha = 0.5):
    size = method[0].size
    elemsSize = N
    h = rhs / N

    y = np.zeros(elemsSize)
    z = np.zeros(elemsSize)
    
    y[0] = InitialСonditions(alpha)[0]
    z[0] = InitialСonditions(alpha)[1]
    
    for i in range(elemsSize - 1):
        ky = np.zeros(size - 1)
        kz = np.zeros(size - 1)
        
        for j in range(size - 1):
            a      = method[j, 1:]
            fstArg = y[i] + method[0, j]
            doty   = np.dot(ky, a)
            dotz   = np.dot(kz, a)
            
            scdArgy = z[i] + h * doty
            scdArgz = z[i] + h * dotz
            
            ky[j] = system(scdArgy, fstArg)[0]
            kz[j] = system(scdArgz, fstArg)[1]

        b      = method[size - 1][1:]
        doty   = np.dot(ky, b)
        dotz   = np.dot(kz, b)

        y[i+1] = y[i] + h * doty
        z[i+1] = z[i] + h * dotz
        
    return y, z
```

```python
RG4 = np.array ([[      0,       0,       0,       0,       0],
                 [1.0 / 2, 1.0 / 2,       0,       0,       0],
                 [1.0 / 2,       0, 1.0 / 2,       0,       0],
                 [      1,       0,       0,       1,       0],
                 [      0, 4.0 / 6, 2.0 / 6, 2.0 / 6, 1.0 / 6]])
```

## x_k = 0.5

```python
xk = 0.5
N  = 10000
y, z = solve(N, xk, RG4, 4.4309)
```

```python
y[-1] # проверка, что при выбранном коэфф. получили правильное значение на правой границе
```

```python
x = np.linspace(0, xk, N)
plt.figure(figsize = [16, 9])
plt.plot(x, y)
plt.title(label = 'xk = 0.5')
plt.xlabel("x")
plt.ylabel("y")
plt.tight_layout() 
plt.grid()
plt.show()
```
