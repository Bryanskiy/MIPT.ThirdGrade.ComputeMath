# Приложение

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_graph(y,z, ltext):
  plt.figure(figsize=[10, 4])
  plt.title(ltext)
  plt.plot(y, z, 'g', label='Фазовая траектория')
  plt.grid()
  plt.xlabel("$y(t)$")
  plt.ylabel("$z(t)$")
  plt.legend(loc='best', bbox_to_anchor=(1, 1))
  plt.tight_layout()
  plt.show()
```

```python
def f1z(z, y, e):
  return e * (1 - y**2) * z - y
def f1y(z,y, e):
  return z
def f2z(z,y, e):
  return -y
def f2y(z, y, e):
  return z - e * ((y**3)/3 - y)

ky = np.zeros(4)
kz = np.zeros(4)
```

##Метод Рунге-Кутты Iго порядка



```python
def rgI(h, e):
  y, z = np.zeros(int(100/h)), np.zeros(int(100/h))
  y[0], z[0] = 2, 0
  for t in range(0, int(100/h) - 1):
    z[t+1] = z[t] + h * f1z(z[t], y[t], e)
    y[t+1] = y[t] + h * f1y(z[t], y[t], e)
  make_graph(y,z, f"Метод Рунге-Кутты Iго порядка, $e = {e}$")
```

```python
def rgI_L(h, e):
  y, z = np.zeros(int(100/h)), np.zeros(int(100/h))
  y[0], z[0] = 2, 0
  for t in range(0, int(100/h) - 1):
    y[t+1] = y[t] + h * f2y(z[t], y[t], e)
    z[t+1] = z[t] + h * f2z(z[t], y[t], e)
  make_graph(y,z, f"Метод Рунге-Кутты Iго порядка, $e = {e}$")
```

##Метод Рунге-Кутты IIго порядка

```python
def rgII(h, e, flag = 0):
  y, z = np.zeros(int(100/h)), np.zeros(int(100/h))
  y[0], z[0], pr = 2, 0, int(100/h) - 1 if flag == 0 else 1
  for t in range(pr):
    kz[0] = f1z(z[t], y[t], e)
    kz[1] = f1z(z[t] + h/2 * kz[0], y[t] + h/2 * ky[0], e)
    ky[0] = f1y(z[t], y[t], e)
    ky[1] = f1y(z[t] + h/2 * kz[0], y[t] + h/2 * ky[0], e)
    y[t+1] = y[t] + ky[1] * h
    z[t+1] = z[t] + kz[1] * h

  if (flag == 0):
    make_graph(y,z, f"Метод Рунге-Кутты IIго порядка, $e = {e}$")
  return y[1], z[1]
```

```python
def rgII_L(h, e, flag = 0):
  y, z = np.zeros(int(100/h)), np.zeros(int(100/h))
  y[0], z[0], pr = 2, 0, int(100/h) - 1 if flag == 0 else 1
  for t in range(pr):
    kz[0] = f2z(z[t], y[t], e)
    kz[1] = f2z(z[t] + h/2 * kz[0], y[t] + h/2 * ky[0], e)
    ky[0] = f2y(z[t], y[t], e)
    ky[1] = f2y(z[t] + h/2 * kz[0], y[t] + h/2 * ky[0], e)
    y[t+1] = y[t] + ky[1] * h
    z[t+1] = z[t] + kz[1] * h

  if (flag == 0):
    make_graph(y,z, f"Метод Рунге-Кутты IIго порядка, $e = {e}$")
  return y[1], z[1]
```

##Метод Рунге-Кутты IIIго порядка

```python
def rgIII(h, e, flag = 0):
  y, z = np.zeros(int(100/h)), np.zeros(int(100/h))
  y[0], z[0], pr = 2, 0, int(100/h) - 1 if flag == 0 else 2
  for t in range(pr):
    kz[0] = f1z(z[t], y[t], e)
    kz[1] = f1z(z[t] + h/3 * kz[0], y[t] + h/3 * ky[0], e)
    kz[2] = f1z(z[t] + 2*h/3 * kz[1], y[t] + 2*h/3 * ky[1], e)
    ky[0] = f1y(z[t], y[t], e)
    ky[1] = f1y(z[t] + h/3 * kz[0], y[t] + h/3 * ky[0], e)
    ky[2] = f1y(z[t] + 2*h/3 * kz[1], y[t] + 2*h/3 * ky[1], e)
    y[t+1] = y[t] + ky[0] * h/4 + ky[2] * 3 * h/4
    z[t+1] = z[t] + kz[0] * h/4 + kz[2] * 3 * h/4

  if (flag == 0):
    make_graph(y,z, f"Метод Рунге-Кутты IIIго порядка, $e = {e}$")
  return y[1], z[1], y[2], z[2]
```

```python
def rgIII_L(h, e, flag = 0):
  y, z = np.zeros(int(100/h)), np.zeros(int(100/h))
  y[0], z[0], pr = 2, 0, int(100/h) - 1 if flag == 0 else 2
  for t in range(pr):
    kz[0] = f2z(z[t], y[t], e)
    kz[1] = f2z(z[t] + h/3 * kz[0], y[t] + h/3 * ky[0], e)
    kz[2] = f2z(z[t] + 2*h/3 * kz[1], y[t] + 2*h/3 * ky[1], e)
    ky[0] = f2y(z[t], y[t], e)
    ky[1] = f2y(z[t] + h/3 * kz[0], y[t] + h/3 * ky[0], e)
    ky[2] = f2y(z[t] + 2*h/3 * kz[1], y[t] + 2*h/3 * ky[1], e)
    y[t+1] = y[t] + ky[0] * h/4 + ky[2] * 3 * h/4
    z[t+1] = z[t] + kz[0] * h/4 + kz[2] * 3 * h/4

  if (flag == 0):
    make_graph(y,z, f"Метод Рунге-Кутты IIIго порядка, $e = {e}$")
  return y[1], z[1], y[2], z[2]
```

##Метод Рунге-Кутты IVго порядка

```python
def rgIV(h, e, flag = 0):
  y, z = np.zeros(int(100/h)), np.zeros(int(100/h))
  y[0], z[0], pr = 2, 0, int(100/h) - 1 if flag == 0 else 3
  for t in range(pr):
    kz[0] = f1z(z[t], y[t], e)
    kz[1] = f1z(z[t] + h/2 * kz[0], y[t] + h/2 * ky[0], e)
    kz[2] = f1z(z[t] + h/2 * kz[1], y[t] + h/2 * ky[1], e)
    kz[3] = f1z(z[t] + h * kz[2], y[t] + h * ky[2], e)
    ky[0] = f1y(z[t], y[t], e)
    ky[1] = f1y(z[t] + h/2 * kz[0], y[t] + h/2 * ky[0], e)
    ky[2] = f1y(z[t] + h/2 * kz[1], y[t] + h/2 * ky[1], e)
    ky[3] = f1y(z[t] + h * kz[2], y[t] + h * ky[2], e)
    y[t+1] = y[t] + ky[0] * h/6 + ky[1] * h/3 + ky[2] * h/3 + ky[3] * h/6
    z[t+1] = z[t] + kz[0] * h/6 + kz[1] * h/3 + kz[2] * h/3 + kz[3] * h/6

  if (flag == 0):
    make_graph(y,z, f"Метод Рунге-Кутты IVго порядка, $e = {e}$")
  return y[1], z[1], y[2], z[2], y[3], z[3]
```

```python
def rgIV_L(h,e, flag = 0):
  y, z = np.zeros(int(100/h)), np.zeros(int(100/h))
  y[0], z[0], pr = 2, 0, int(100/h) - 1 if flag == 0 else 3
  for t in range(pr):
    kz[0] = f2z(z[t], y[t],e)
    kz[1] = f2z(z[t] + h/2 * kz[0], y[t] + h/2 * ky[0], e)
    kz[2] = f2z(z[t] + h/2 * kz[1], y[t] + h/2 * ky[1], e)
    kz[3] = f2z(z[t] + h * kz[2], y[t] + h * ky[2], e)
    ky[0] = f2y(z[t], y[t], e)
    ky[1] = f2y(z[t] + h/2 * kz[0], y[t] + h/2 * ky[0], e)
    ky[2] = f2y(z[t] + h/2 * kz[1], y[t] + h/2 * ky[1], e)
    ky[3] = f2y(z[t] + h * kz[2], y[t] + h * ky[2], e)
    y[t+1] = y[t] + ky[0] * h/6 + ky[1] * h/3 + ky[2] * h/3 + ky[3] * h/6
    z[t+1] = z[t] + kz[0] * h/6 + kz[1] * h/3 + kz[2] * h/3 + kz[3] * h/6

  if (flag == 0):
    make_graph(y,z, f"Метод Рунге-Кутты IVго порядка, $e = {e}$")
  return y[1], z[1], y[2], z[2], y[3], z[3]
```

##Метод Адамса IIго порядка

```python
def amII(h, e):
  y, z = np.zeros(int(100/h)), np.zeros(int(100/h))
  y[0], z[0], y[1], z[1] = 2, 0, *rgII(h, e, 1)
  for t in range(1, int(100/h) - 1):
    y[t+1] = y[t] + h * (3/2 * f1y(z[t], y[t], e) - 1/2 * f1y(z[t-1], y[t-1], e))
    z[t+1] = z[t] + h * (3/2 * f1z(z[t], y[t], e) - 1/2 * f1z(z[t-1], y[t-1], e))
  make_graph(y,z, f"Метод Адамса IIго порядка, $e = {e}$")
```

```python
def amII_L(h, e):
  y, z = np.zeros(int(100/h)), np.zeros(int(100/h))
  y[0], z[0], y[1], z[1] = 2, 0, *rgII_L(h, e, 1)
  for t in range(1, int(100/h) - 1):
    y[t+1] = y[t] + h * (3/2 * f2y(z[t], y[t],e) - 1/2 * f2y(z[t-1], y[t-1],e))
    z[t+1] = z[t] + h * (3/2 * f2z(z[t], y[t],e) - 1/2 * f2z(z[t-1], y[t-1],e))
  make_graph(y,z, f"Метод Адамса IIго порядка, $e = {e}$")
```

##Метод Адамса IIIго порядка

```python
def amIII(h, e):
  y, z = np.zeros(int(100/h)), np.zeros(int(100/h))
  y[0], z[0], y[1], z[1], y[2], z[2] = 2,0, *rgIII(h,e, 1)
  for t in range(2, int(100/h) - 1):
    y[t+1] = y[t] + h * (23/12 * f1y(z[t], y[t],e) - 16/12 * f1y(z[t-1], y[t-1],e) + 5/12 * f1y(z[t-2], y[t-2],e))
    z[t+1] = z[t] + h * (23/12 * f1z(z[t], y[t],e) - 16/12 * f1z(z[t-1], y[t-1],e) + 5/12 * f1z(z[t-2], y[t-2],e))
  make_graph(y,z, f"Метод Адамса IIIго порядка, $e = {e}$")
```

```python
def amIII_L(h, e):
  y, z = np.zeros(int(100/h)), np.zeros(int(100/h))
  y[0], z[0], y[1], z[1], y[2], z[2] = 2, 0, *rgIII_L(h,e, 1)
  for t in range(2, int(100/h) - 1):
    y[t+1] = y[t] + h * (23/12 * f2y(z[t], y[t],e) - 16/12 * f2y(z[t-1], y[t-1],e) + 5/12 * f2y(z[t-2], y[t-2],e))
    z[t+1] = z[t] + h * (23/12 * f2z(z[t], y[t],e) - 16/12 * f2z(z[t-1], y[t-1],e) + 5/12 * f2z(z[t-2], y[t-2],e))
  make_graph(y,z, f"Метод Адамса IIIго порядка, $e = {e}$")
```

##Метод Адамса IVго порядка

```python
def amIV(h, e):
  y, z = np.zeros(int(100/h)), np.zeros(int(100/h))
  y[0], z[0], y[1], z[1], y[2], z[2], y[3], z[3] = 2, 0, *rgIV(h, e, 1)
  for t in range(3, int(100/h) - 1):
    y[t+1] = y[t] + h * (55/24 * f1y(z[t], y[t],e) - 59/24 * f1y(z[t-1], y[t-1],e) + 37/24 * f1y(z[t-2], y[t-2],e) - 9/24 * f1y(z[t-3], y[t-3],e))
    z[t+1] = z[t] + h * (55/24 * f1z(z[t], y[t],e) - 59/24 * f1z(z[t-1], y[t-1],e) + 37/24 * f1z(z[t-2], y[t-2],e) - 9/24 * f1z(z[t-3], y[t-3],e))
  make_graph(y,z, f"Метод Адамса IVго порядка, $e = {e}$")
```

```python
def amIV_L(h, e):
  y, z = np.zeros(int(100/h)), np.zeros(int(100/h))
  y[0], z[0], y[1], z[1], y[2], z[2], y[3], z[3] = 2, 0, *rgIV_L(h, e, 1)
  for t in range(3, int(100/h) - 1):
    y[t+1] = y[t] + h * (55/24 * f2y(z[t], y[t],e) - 59/24 * f2y(z[t-1], y[t-1],e) + 37/24 * f2y(z[t-2], y[t-2],e) - 9/24 * f2y(z[t-3], y[t-3],e))
    z[t+1] = z[t] + h * (55/24 * f2z(z[t], y[t],e) - 59/24 * f2z(z[t-1], y[t-1],e) + 37/24 * f2z(z[t-2], y[t-2],e) - 9/24 * f2z(z[t-3], y[t-3],e))
  make_graph(y,z, f"Метод Адамса IVго порядка, $e = {e}$")
```

```python
def graph_all(h, e):
  rgI(h,e)
  rgII(h,e)
  rgIII(h,e)
  rgIV(h,e)
  amII(h,e)
  amIII(h,e)
  amIV(h,e)

def graph_all_L(h, e):
  rgI_L(h,e)
  rgII_L(h,e)
  rgIII_L(h,e)
  rgIV_L(h,e)
  amII_L(h,e)
  amIII_L(h,e)
  amIV_L(h,e)
```
